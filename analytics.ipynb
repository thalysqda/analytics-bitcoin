O que é bitcoin?

O bitcoin é uma moeda virtual – a primeira criada no mundo – e pode ser usado para a compra de serviços, produtos e quaisquer outros itens em estabelecimentos que aceitem ser pagos com ele.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as Dash
```

```python
df = pd.read_excel('Dados_Bitcoin.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
plt = px.line(df, x='Date', y='Close', title='Preço do Bitcoin ao longo do tempo')
plt.show()
```

```python
df.set_index('Date', inplace=True)
```

```python
media_movel = df['Close'].rolling(5).mean()
media_tendencia = df['Close'].rolling(20).mean()
```

```python
# criar um dashboard

figura = Dash.Figure()

# Adcionando primeiro eixo
figura.add_trace(
    Dash.Scatter(
        x = df.index,
        y = df.Close,
        mode = 'lines',
        name = 'Fechamento',
        marker_color = '#ff7f0e',
        opacity = 0.5,
    )
)

# Adcionando a média
figura.add_trace(
    Dash.Scatter(
        x = df.index,
        y = media_movel,
        mode  ='lines',
        name = 'Média Móvel',
        marker_color = '#2ca02c',
    )
)

# Adcionando a tendência
figura.add_trace(
    Dash.Scatter(
        x = df.index,
        y = media_tendencia,
        mode  ='lines',
        name = 'Tendência',
        marker_color = '#d62728',
    )
)

# Ajustes no layout
figura.update_layout(
    
    title=dict(text='Análise do fechamento Bitcoin', font=dict(size=20)),

    # Ajustando eixo x.
    xaxis = dict(
        title=dict(text='Período Histórico', font=dict(size=14)),
        tickfont=dict(size=10),
    ),

# Ajustando eixo y.
yaxis = dict(
    title=dict(text='Preço fechamento ($)', font=dict(size=14)),
    tickfont=dict(size=10),
),

# Parametros para Legenda
    legend = dict (
        y=1,
        x=1,
        bgcolor='rgba(255, 255, 255, 1)',
        bordercolor='rgba(255, 255, 255, 1)'
    )   
)
```
